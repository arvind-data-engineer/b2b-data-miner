 This notebook reads `data/cleaned_products.csv`, which is produced by `scraper/crawler.py`, and explores category coverage, supplier geography, prices, keywords, and missing-data gaps.

```python
from pathlib import Path
import re
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd

DATA_PATH = Path('..') / 'data' / 'cleaned_products.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df['category'].value_counts().plot(kind='barh', title='Products by category')
plt.xlabel('Products')
plt.show()
```

```python
prices = pd.to_numeric(df.get('price_min_inr'), errors='coerce').dropna()
prices.describe()
```

```python
if not prices.empty:
    prices.plot(kind='hist', bins=20, title='Price distribution: minimum listed INR')
    plt.xlabel('Minimum listed price')
    plt.show()
```

```python
df['state'].dropna().value_counts().head(10).plot(kind='barh', title='Top supplier states')
plt.xlabel('Suppliers')
plt.show()
```

```python
stopwords = {'and', 'for', 'with', 'the', 'machine', 'machines', 'system', 'plant'}
words = []
for name in df['product_name'].dropna():
    words.extend(w for w in re.findall(r'[a-zA-Z]{3,}', name.lower()) if w not in stopwords)
pd.Series(dict(Counter(words).most_common(20))).plot(kind='barh', title='Frequent product keywords')
plt.xlabel('Mentions')
plt.show()
```
